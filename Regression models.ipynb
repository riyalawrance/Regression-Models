<h1>Linear Regression</h1>

```python

# Importing the libraries

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Importing the datasets
 
datasets = pd.read_csv('Data.csv')

X = datasets.iloc[:, :-1].values
Y = datasets.iloc[:, -1].values

# Splitting the dataset into the Training set and Test set

X_Train, X_Test, Y_Train, Y_Test = train_test_split(X, Y, test_size = 0.4, random_state = 2)

# Fitting Simple Linear Regression to the training set


regressor = LinearRegression()
regressor.fit(X_Train, Y_Train)

# Predicting the Test set result ï¿¼

Y_Pred = regressor.predict(X_Test)

r2_1 = r2_score(Y_Test,Y_Pred)
print("Coefficient of Determination R^2 =",r2_1)

sse_1 = np.sum((Y_Test - Y_Pred)**2)
print("SSE =",sse_1)

# Visualising the Training set results

plt.scatter(X_Train, Y_Train, color = 'red')
plt.plot(X_Train, regressor.predict(X_Train), color = 'blue')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

# Visualising the Test set results

plt.scatter(X_Test, Y_Test, color = 'red')
plt.plot(X_Test, regressor.predict(X_Test), color = 'blue')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

<h1>Polynomial Regression with Degree 2</h1>

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

#importing datasets
datasets = pd.read_csv('Data.csv')
X = datasets.iloc[:,:-1].values
Y = datasets.iloc[:,-1].values

#transforming input data
transformer = PolynomialFeatures(degree = 2)
X_ = transformer.fit_transform(X)

#creating a model
model = LinearRegression()
model.fit(X_,Y)

#Predicting the result
Y_Pred = model.predict(X_)
r2_2 = r2_score(Y,Y_Pred)
print("Coefficient of Determination R^2 =",r2_2)

sse_2 = np.sum((Y - Y_Pred)**2)
print("SSE =",sse_2)

#showing the model
xGrid = np.arange(min(X), max(X), 0.1)
xGrid = xGrid.reshape((len(xGrid), 1))
plt.scatter(X, Y, color = 'red')
plt.plot(xGrid, model.predict(transformer.fit_transform(xGrid)), color = 'blue')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

<h1>Polynomial Regression with Degree 3</h1>

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

#importing datasets
datasets = pd.read_csv('Data.csv')
X = datasets.iloc[:,:1].values
Y = datasets.iloc[:,-1].values

#transforming input data
transformer = PolynomialFeatures(degree = 3)
transformer.fit(X)
X_ = transformer.transform(X)

#creating a model
model = LinearRegression()
model.fit(X_,Y)

#Predicting the result
Y_Pred = model.predict(X_)
r2_3 = r2_score(Y,Y_Pred)
print("Coefficient of Determination R^2 =",r2_3)

sse_3 = np.sum((Y - Y_Pred)**2)
print("SSE =",sse_3)

#showing the model
xGrid = np.arange(min(X), max(X), 0.1)
xGrid = xGrid.reshape((len(xGrid), 1))
plt.scatter(X, Y, color = 'red')
plt.plot(xGrid, model.predict(transformer.fit_transform(xGrid)), color = 'blue')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
